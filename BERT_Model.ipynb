```python
import pandas as pd
import ktrain
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from ktrain import text
```

```python
df = pd.read_csv("/content/final_data1.csv")
df
```

```python
x_train, x_test, y_train, y_test = train_test_split(df['Hindi'], df['label'], test_size=0.2, random_state=0)
```

```python
t = text.Transformer("monsoon-nlp/hindi-bert", maxlen=500, class_names=list(set(y_train.values)))
```

```python
#preprocessing

trn = t.preprocess_train(x_train.to_numpy(), y_train.to_numpy())
tst = t.preprocess_test(x_test.to_numpy(), y_test.to_numpy())
```

```python
#model training

model = t.get_classifier()
learner = ktrain.get_learner(model, train_data=trn, val_data=tst, batch_size=6)
```

```python
lr = 1e-3
epochs = 10

history = learner.fit_onecycle(lr, epochs)
```

```python
# prediction
pred = ktrain.get_predictor(learner.model, preproc=t)
```

```python
i = 55
print('Actual Text = ', x_test.iloc[i])
print('\n')
print("*"*50)
print("Predicted Label = ",pred.predict(x_test.iloc[i]))
print('Actual Label = ',y_test.iloc[i])
```
